### Горные хребты

```python
import pandas as pd
import numpy as np
```

```python
#Читаем данные из файла
data = pd.read_csv( 'Mountains.csv', delimiter = ',')
#data.info()
```

```python
#Смотрим что прочиталось
data.head(10)
```

```python
#Отбираем для работы только 2 столбца, гора и горный хребет, которому гора принадлежит
dataselected=data.loc[0:,['Mountain', 'Parent mountain']]
dataselected.head()
```

```python
#Группируем горы по хребтам. Считаем количество гор в каждом хребте
dataselected=dataselected.groupby('Parent mountain').count()
dataselected.rename( columns={'Mountain': 'Count_of_mountains'}, inplace=True)
dataselected.head()
```

```python
#Задаем маску: выбираем только те хребты, где больше 3 гор
mask = ( dataselected.Count_of_mountains >3 )
```

```python
#Выводим названия хребтов и количество входящих в них гор
dataselected.loc[ mask ]
```
