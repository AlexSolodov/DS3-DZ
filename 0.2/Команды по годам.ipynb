### Команды по годам

```python
import pandas as pd
import numpy as np
```

```python
#Читаем данные из файла
data = pd.read_csv( 'teams_stats.csv', delimiter = ',')
#data.info()
```

```python
#Смотрим что прочиталось
data.head(10)
```

```python
#data.info()
```

```python
#Отбираем для работы только 2 столбца, Год и команда
dataselected=data.loc[0:,['Year', 'Tm']]
dataselected.head()
```

```python
#Группируем команды по годам, берем только уникальные значения
d2=dataselected.groupby('Year').nunique('Tm')
```

```python
#Удаляем лишний столбец Year
del d2 ['Year']
d2.rename( columns = {'Tm': 'Teams per season'}, inplace=True)
d2
```
