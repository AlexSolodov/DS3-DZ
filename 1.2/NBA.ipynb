```python
# Python 2 and 3 compatibility
# pip install future
from __future__ import (absolute_import, division,
                        print_function, unicode_literals)
# отключим предупреждения Anaconda
import warnings
warnings.simplefilter('ignore')

# будем отображать графики прямо в jupyter'e
%pylab inline
#графики в svg выглядят более четкими
%config InlineBackend.figure_format = 'svg' 

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 6,3
import pandas as pd
import seaborn as sns
```

```python
#Загружаем данные со статистикой игроков
df_Seasons_stats = pd.read_csv('Seasons_Stats.csv')
print(df_Seasons_stats.shape)
```

```python
#df_Seasons_stats.info()
```

```python
#Смотрим, что загрузилось
df_Seasons_stats.head()
```

```python
#Из загруженных данных отфильтровываем только информацию о годах и командах
dataselected=df_Seasons_stats.loc[0:,['Year', 'Tm']]
dataselected.head()
```

```python
#Группируем команды по годам, берем только уникальные значения
d2=dataselected.groupby('Year').nunique('Tm')
```

```python
#Удаляем лишний столбец Year
del d2 ['Year']
d2.rename( columns = {'Tm': 'Teams per season'}, inplace=True)
#d2
```

```python
#Выводим простейший график:"Количество команд участвовавших в чемпионате НБА по годам"
ax=d2.plot()
ax.set_title('Number of teams by years')
ax.set_xlabel('Years')
ax.set_ylabel('Number of teams')
```

```python
#Отфильтруем колонки с которыми в дальнейшем будем работать
useful_cols = ['Year', 'Player', 'Pos', 'Age', 
               'Tm', 'G', 'FG',
               'AST', 'PF', 'PTS'
              ]
df2 = df_Seasons_stats[useful_cols]
```

```python
df2.tail(20)
```

```python
df2.info()
```

```python
#Значений NaN в нашей выборке совсем не много. Удаляем их, чтобы не мешали в проведении экспериментов
df2 = df2.dropna()
print(df2.shape)
```

```python
#В загруженных данных информация о годах, проведенных матчах, передачах и набранных очках имеет тип
#float, что не всегда удобно, преобразуем ее к типу int
df2['Year'] = df2.Year.astype('int32')
df2['G'] = df2.G.astype('int32')
df2['AST'] = df2.AST.astype('int32')
df2['PTS'] = df2.PTS.astype('int32')
```

```python
df2.tail(10)
```

```python
#Посмотрим, какие команды участвовали в чемпионате чаще всего
dtm=dataselected.groupby('Tm').nunique('Year')
```

```python
mask=(dtm.Year>50)
```

```python
dtm2=dtm.loc[mask].head(10)
del dtm2 ['Tm']
dtm2
```

```python
#Создадим маску по которой в нашей выборке останутся только команды участвовавшие чаще других
mask=((df2.Tm =='BOS') | (df2.Tm =='CHI') | (df2.Tm =='DET') | (df2.Tm =='LAL') 
      | (df2.Tm =='NYK') | (df2.Tm =='PHI'))
```

```python
df3=df2.loc[mask]
```

```python
us_cols = ['Year', 'Tm', 'PTS']
```

```python
#Создадим сводную таблицу на осих Х- отложим года чемпионатов, на оси Y- отобранные команды
#а в качестве значений укажем количество набранных очков
year_team_points = df3[us_cols].pivot_table(
                        index='Year', 
                        columns='Tm', 
                        values='PTS', 
                        aggfunc=sum).fillna(0).applymap(float)
```

```python
#Немоного изменим параметры вывода графиков
rcParams['figure.figsize'] = 9,5
```

```python
#Нарисуем нашу сводную таблицу со значениями по-умолчанию
year_team_points.plot()
```

```python
#На графике видно, когда какие команды начинали играть в НБА, а так же два характерных минимума.
#Минимальное количество очков команды набирали в годы локаутов
```

```python
#Отобразим те же данные при помощи другого графике. например bar char. Для этого опять немного изменим
#параметры отображения
rcParams['figure.figsize'] = 20,8
```

```python
year_team_points.plot(kind='bar', rot=45);
```

## <center>Часть 3. Plotly

```python
#Импортируем Plotly
from plotly.offline import init_notebook_mode, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
#Построим график набранных очков и сделанных передач по годам
#Для начала сформируем таблицу набранных очков
pts_years_df = df2.groupby('Year')[['PTS']].sum()
pts_years_df.head()
```

```python
#Теперь, таблицу сделанных передач
ast_years_df = df2.groupby('Year')[['AST']].sum()
ast_years_df.head()
```

```python
#С джойним 2 таблицы
years_df = pts_years_df.join(ast_years_df)
years_df.head()
```

В `plotly` строится визуализация объекта `Figure`, который состоит из данных (массив линий, которые в библиотеке называются `traces`) и оформления/стиля, за который отвечает объект `layout`. В простых случаях можно вызывать функцию `iplot` и просто от массива `traces`.

```python
#Строим графики
trace0 = go.Scatter(
    x=years_df.index,
    y=years_df.PTS,
    name='Points'
)

trace1 = go.Scatter(
    x=years_df.index,
    y=years_df.AST,
    name='Asists'
)

data = [trace0, trace1]
layout = {'title': 'Statistics of points and asists'}

fig = go.Figure(data=data, layout=layout)

iplot(fig, show_link=False)
```

Также можно сразу сохранить график в виде html-файла.

```python
plotly.offline.plot(fig, filename='years_stats.html', show_link=False);
```
