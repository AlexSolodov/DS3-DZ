### Прогноз размещения на площадке (отрасли + регионы закупок) 

```python
import pandas as pd
import numpy as np
%matplotlib inline

from sklearn.metrics import mean_squared_log_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from lightgbm import LGBMRegressor
```

```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'sans-serif'
rcParams['font.sans-serif'] = ['DejaVu Sans']
#Немоного изменим параметры вывода графиков
rcParams['figure.figsize'] = 18,10
```

```python
data = pd.read_excel('ТоварыРегионыНедели.xlsx')
data.head()
```

```python
data.info()
```

```python
columns=['Область', 'Регион', 'Obl2']
data.drop(columns, axis=1, inplace=True)
```

```python
melt = data.melt(id_vars='UnKod', var_name='Week', value_name='Proc')

melt['Week'] = melt['Week'].str.extract('(\d+)', expand=False).astype(int)

melt = melt.sort_values(['UnKod', 'Week'])
melt.head()
```

```python
melt2 =  melt.copy()
melt2['Last_Week_Proc'] = melt2.groupby(['UnKod'])['Proc'].shift()
melt2['Last_Week_Diff'] = melt2.groupby(['UnKod'])['Last_Week_Proc'].diff()
melt2 = melt2.dropna()
melt2.head()
```

```python
melt2.info()
```

```python
def rmsle(ytrue, ypred):
    return np.sqrt(mean_squared_log_error(ytrue, ypred))
```

```python
mean_error = []
total_error = []
Proc = []
predict_Proc = []
for week in range(128,140):
    train = melt2[melt2['Week'] < week]
    val = melt2[melt2['Week'] == week]
    yts=val['Proc']
    
    p = val['Last_Week_Proc'].values
    
    total_yts=sum(yts)
    total_p=sum(p)
    
    error = rmsle(yts, p)
    abs_error = abs(total_yts-total_p)

    print('Week %d - Error %.5f, Real Total %.1f - Predict Total %.1f, Abs Error %.1f' % (week, error, total_yts, total_p, abs_error))
    mean_error.append(error)
    total_error.append(abs_error)
    Proc.append(total_yts)
    predict_Proc.append(total_p)
print('Mean Error = %.5f' % np.mean(mean_error))
print('Mean Abs Error = %.5f' % np.mean(total_error))
```

```python
plt.plot(Proc)
plt.plot(predict_Proc)
plt.title('Procedures (week 128 - 140)')
plt.xlabel('Week numbers')
plt.ylabel('Procedures count')
plt.legend(('Real', 'Predict'))
```
