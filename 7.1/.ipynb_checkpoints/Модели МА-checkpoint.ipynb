### Привидение рядов к стационарному виду

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
import statsmodels.api as sm
import statsmodels.tsa.api as smt
import matplotlib.pyplot as plt
from scipy.stats import boxcox
%matplotlib inline
```

```python
def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)
```

```python
def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)

        plt.tight_layout()
    return
```

### Индекс Доу-Джонса

Загрузим график изменения индекса Доу-Джонса и приведем его к стационарности

```python
series = pd.read_csv("weekly-closings-of-the-dowjones-.csv", sep=";")['Count']
```

```python
series.head()
```

```python
tsplot(series)
```

Как мы видим, и тест Дики-Фуллера и графики коррелограмм не отвергают гипотезу о нестационарности ряда. Особых изменений дисперсии не видно, но видно несколько трендов, попробуем убрать их дифференцированием

```python
series2 = series[1:]
series3 = series[:-1]
```

```python
len(series)
```

```python
for i in range(len(series)-1):
    series[i]=series2[i+1]-series3[i]
```

```python
series.tail()
```

```python
tsplot(series[:-1])
```

Теперь тест Дики-Фуллера и графики коррелограмм отвергают гипотезу о нестационарности ряда!

### Продажи домов

Загрузим данные о продажах домов в США

```python
series_2 = pd.read_csv("sales-of-new-onefamily-houses-us.csv", sep=";")['Count']
```

```python
series_2.head()
```

```python
tsplot(series_2)
```

Как мы видим, и тест Дики-Фуллера и графики коррелограмм не отвергают гипотезу о нестационарности ряда. Для начала уберем изменение дисперсии при помощи преобразования Бокса-Кокса

```python
series_2 = boxcox(series_2, 0)
```

```python
tsplot(series_2)
```

Стало еще лучше, но по графику коррелограммы видно сильное влияние сезонности. Уберем ее

```python
series_2 = series_2[12:] - series_2[:-12]
```

```python
tsplot(series_2)
```

Стало лучше, но тест Дикки-Фуллера все еще не отвергает гипотезу о нестационарности ряда. По графику ряда видно наличие сильного тренда. Уберем его дифференцированием.

```python
series_2 = series_2[1:] - series_2[:-1]
```

```python
tsplot(series_2)
```

Теперь тест Дики-Фуллера и графики коррелограмм отвергают гипотезу о нестационарности ряда!
